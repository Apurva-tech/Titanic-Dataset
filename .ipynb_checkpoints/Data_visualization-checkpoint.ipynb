```python
#DPP
import pandas as pd
import numpy as np
import random as rnd
#visualisation
import matplotlib.pyplot as plt
import seaborn as sns
#MODEL
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC,LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
```

```python
train_df = pd.read_csv('train.csv')
print(train_df.shape)
print(train_df.count())
```

```python
fig = plt.figure(figsize=(18,6))
train_df.Survived.value_counts(normalize=True).plot(kind='bar',alpha=0.5)
plt.show()
plt.hist(train_df.Survived)
plt.winter()
plt.show()
plt.scatter(train_df.Survived,train_df.Age)

plt.title("Age vs Survived")

fig = plt.figure(figsize=(18,6))
train_df.Pclass.value_counts(normalize=True).plot(kind='bar',alpha=0.5)
plt.subplot2grid((2,3),(1,0),colspan=2)
plt.show()
plt.scatter(train_df.Pclass,train_df.Survived)
plt.show()
```

```python
for s in [1,2,3]:
    train_df.Age[train_df.Pclass == s].plot(kind='kde')
plt.title("Age wrt class")
plt.legend(('1st','2nd','3rd'))
plt.show()
train_df.Embarked.value_counts(normalize=True).plot(kind='bar',alpha=0.5)   
plt.show()
```

```python
#visualise gender
fig = plt.figure(figsize=(18,6))
plt.subplot2grid((3,4),(0,0))
train_df.Survived.value_counts(normalize=True).plot(kind='bar',alpha=0.5)
plt.title("Survived")
plt.show()

fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(0,0))
train_df.Survived[train_df.Sex == 'male'].value_counts(normalize=True).plot(kind='bar',alpha=0.5)
plt.title("Male survivors")
plt.show()

fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(0,0))
train_df.Survived[train_df.Sex == 'female'].value_counts(normalize=True).plot(kind='bar',alpha=0.5,color='y')
plt.title("Female Survivor")
plt.show()


fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(0,0))
train_df.Sex[train_df.Survived == 1].value_counts(normalize=True).plot(kind='bar',alpha=0.5,color='b')
plt.title("Sex survived")
plt.show()


```

```python
plt.subplot2grid((2,3),(1,0),colspan=4)
for s in [1,2,3]:
    train_df.Survived[train_df.Pclass == s].plot(kind='kde')
plt.title("Survived wrt class")
plt.legend(('1st','2nd','3rd'))
plt.show()
```

```python
fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(2,0))
train_df.Survived[(train_df.Sex == 'male') & (train_df.Pclass ==1)].value_counts(normalize=True).plot(kind='bar',alpha=0.5)
plt.title("rich Men survivors")
plt.show()


fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(2,0))
train_df.Survived[(train_df.Sex == 'male') & (train_df.Pclass ==3)].value_counts(normalize=True).plot(kind='bar',alpha=0.5)
plt.title("poor Men survivors")
plt.show()

fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(2,0))
train_df.Survived[(train_df.Sex == 'female') & (train_df.Pclass ==1)].value_counts(normalize=True).plot(kind='bar',alpha=0.5,color='m')
plt.title("rich women survivors")
plt.show()

fig = plt.figure(figsize=(15,6))
plt.subplot2grid((3,4),(2,0))
train_df.Survived[(train_df.Sex == 'female') & (train_df.Pclass ==3)].value_counts(normalize=True).plot(kind='bar',alpha=0.5,color='m')
plt.title("poor women survivors")
plt.show()
```
